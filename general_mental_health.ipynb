```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("synthetic_mental_health_dataset.csv")
df.info()
df.describe()
```

```python
print("Missing values:\n", df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.head()
```

```python
df.tail()
```

```python
X = df.drop(columns=['stress_level'])
y = df['stress_level']
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# sir's method
# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Scores
train_score = r2_score(y_train, y_train_pred)
test_score = r2_score(y_test, y_test_pred)
print("Training R² Score:", train_score)
print("Testing  R² Score:", test_score)

total_score = ((train_score - test_score)*100)
print("total:",total_score)
```

```python
# Define the model
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

# Train the model
rf_model.fit(X_train, y_train)

# Predict
y_pred_rf = rf_model.predict(X_test)

# Evaluation
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print("RMSE:", rmse_rf)
print("R2 Score:", r2_rf)
```

```python
from sklearn.metrics import r2_score
rf_train_pred = rf_model.predict(X_train)
rf_test_pred = rf_model.predict(X_test)
print("Train R²:", r2_score(y_train, rf_train_pred))
print("Test  R²:", r2_score(y_test, rf_test_pred))
```
